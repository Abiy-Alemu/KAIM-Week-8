```python
import pandas as pd

# Assuming df_f is your DataFrame
# Define a function to remove outliers based on IQR
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)  # 25th percentile
    Q3 = df[column].quantile(0.75)  # 75th percentile
    IQR = Q3 - Q1  # Interquartile range

    # Determine lower and upper bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Filter the DataFrame to remove outliers
    filtered_df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return filtered_df

# Remove outliers for 'purchase_value'
df_filtered_purchase_value = remove_outliers_iqr(df_f, 'purchase_value')

# Remove outliers for 'age'
df_filtered = remove_outliers_iqr(df_filtered_purchase_value, 'age')

# Display the shape of the original and filtered DataFrames
print("Original DataFrame shape:", df_f.shape)
print("Filtered DataFrame shape:", df_filtered.shape)

# Optional: Check for remaining outliers in the filtered DataFrame
print("Remaining outliers in purchase_value:",
      df_filtered[~df_filtered['purchase_value'].between(df_filtered['purchase_value'].quantile(0.25) - 1.5 * (df_filtered['purchase_value'].quantile(0.75) - df_filtered['purchase_value'].quantile(0.25)),
                                                       df_filtered['purchase_value'].quantile(0.75) + 1.5 * (df_filtered['purchase_value'].quantile(0.75) - df_filtered['purchase_value'].quantile(0.25)))].shape[0])

print("Remaining outliers in age:",
      df_filtered[~df_filtered['age'].between(df_filtered['age'].quantile(0.25) - 1.5 * (df_filtered['age'].quantile(0.75) - df_filtered['age'].quantile(0.25)),
                                               df_filtered['age'].quantile(0.75) + 1.5 * (df_filtered['age'].quantile(0.75) - df_filtered['age'].quantile(0.25)))].shape[0])
```

## 4. Merge Datasets for Geolocation Analysis

```python
import numpy as np

# Ensure 'ip_address_int' is of integer type, replacing non-integer values with NaN
df_f['ip_address_int'] = pd.to_numeric(df_f['ip_address_int'], errors='coerce').fillna(0).astype(np.int64)

# Convert 'lower_bound_ip_address' and 'upper_bound_ip_address' to integer type in df_I
df_I['lower_bound_ip_address'] = pd.to_numeric(df_I['lower_bound_ip_address'], errors='coerce').fillna(0).astype(np.int64)
df_I['upper_bound_ip_address'] = pd.to_numeric(df_I['upper_bound_ip_address'], errors='coerce').fillna(0).astype(np.int64)

# Now proceed with the merge
merged_data = pd.merge(df_f, df_I, 
                       left_on='ip_address_int', 
                       right_on='lower_bound_ip_address',
                       how='left')

# Drop unnecessary columns after merging
merged_data.drop(['ip_address_int', 'lower_bound_ip_address', 'upper_bound_ip_address'], axis=1, inplace=True)

merged_data.head()
```

## 5. Feature Engineering
#1.Transaction Frequency and Velocity Calculation

```python
# Assume 'user_id' is used for frequency calculation
# Calculate daily transaction frequency for each user
merged_data['transaction_date'] = pd.to_datetime(merged_data['purchase_time']).dt.date
transaction_frequency = merged_data.groupby(['user_id', 'transaction_date']).size().reset_index(name='daily_transaction_count')

# Merge the frequency back into the original dataset
merged_data = pd.merge(merged_data, transaction_frequency, on=['user_id', 'transaction_date'], how='left')

# Calculate transaction velocity (transactions per hour)
merged_data['hour_of_day'] = pd.to_datetime(merged_data['purchase_time']).dt.hour
hourly_transaction_velocity = merged_data.groupby(['user_id', 'hour_of_day']).size().reset_index(name='hourly_transaction_count')

# Merge the velocity back into the original dataset
merged_data = pd.merge(merged_data, hourly_transaction_velocity, on=['user_id', 'hour_of_day'], how='left')

merged_data.head()
```

#2.Time-Based Features

```python
# Extract time-based features
merged_data['hour_of_day'] = pd.to_datetime(merged_data['purchase_time']).dt.hour
merged_data['day_of_week'] = pd.to_datetime(merged_data['purchase_time']).dt.dayofweek

# Optional: Calculate additional time-based features
merged_data['month_of_year'] = pd.to_datetime(merged_data['purchase_time']).dt.month
merged_data['is_weekend'] = merged_data['day_of_week'].apply(lambda x: 1 if x >= 5 else 0)

merged_data.head()
```
